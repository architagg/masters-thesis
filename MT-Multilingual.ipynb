```python
def is_english_subfolder(folder_name):
    return 'En' in folder_name
```

```python
input_base_path = "/content/drive/MyDrive/thesis-data/Aligned_Articles-LASER2-new"
output_base_path = "/content/drive/MyDrive/thesis-data/Translated_Articles-LASER2-final"
if not os.path.exists(output_base_path):
    os.makedirs(output_base_path)
```

```python
for subfolder in os.listdir(input_base_path):
    subfolder_path = os.path.join(input_base_path, subfolder)
    if os.path.isdir(subfolder_path):
        translated_subfolder_path = os.path.join(output_base_path, subfolder)
        os.makedirs(translated_subfolder_path, exist_ok=True)

        for filename in os.listdir(subfolder_path):
            if filename.endswith("_aligned.txt"):
                with open(os.path.join(subfolder_path, filename), "r", encoding="utf-8") as file:
                    lines = file.readlines()

                output_filename = os.path.join(translated_subfolder_path, filename.replace("_aligned.txt", "_translated.txt"))
                with open(output_filename, "w", encoding="utf-8") as out_file:
                    for line in lines:
                        if "---->" in line:
                            parts = line.split("---->")
                            sentence1 = parts[0].strip()
                            sentence2 = parts[1].strip()

                            if is_english_subfolder(subfolder):
                                sentence2_translated = translate_text(sentence2)
                                out_file.write(f"{sentence1} ----> {sentence2_translated}\n")
                            else:
                                sentence1_translated = translate_text(sentence1)
                                sentence2_translated = translate_text(sentence2)
                                out_file.write(f"{sentence1_translated} ----> {sentence2_translated}\n")
```

```python
def translate_text(text, target_language="en"):
    """Translate the given text to the target language."""
    if not text:
        return ''
    result = translate_client.translate(text, target_language=target_language)
    return result["translatedText"]
```

```python
def translate_files_in_folder(input_folder, output_folder):
    """Translate all text files in the given input folder."""
    for filename in os.listdir(input_folder):
        if filename.endswith(".txt"):
            input_file_path = os.path.join(input_folder, filename)
            output_file_path = os.path.join(output_folder, filename)

            # Read the content of the file
            with open(input_file_path, "r", encoding="utf-8") as file:
                text = file.read()

            # Translate the text
            translated_text = translate_text(text)

            # Write the translated text to the output file
            with open(output_file_path, "w", encoding="utf-8") as file:
                file.write(translated_text)
            print(f"Translated {input_file_path} to {output_file_path}")
```
