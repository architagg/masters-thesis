```python
import requests
from bs4 import BeautifulSoup

# Function to filter out 'p' tags inside tables
def not_in_table(tag):
    return tag.name == 'p' and not any(ancestor.name == 'table' for ancestor in tag.find_parents())

# Fetch Wikipedia page content
response = requests.get('https://en.wikipedia.org/wiki/Russo-Ukrainian_War')
if response.status_code != 200:
    print("Failed to get content.")
else:
    # Parse HTML
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find the div with id='bodyContent'
    body_content = soup.find('div', {'id': 'bodyContent'})
    
    # Initialize summary text
    summary_text = ''
    
    # Loop through the content to find 'p' tags before the first 'h2' within the 'bodyContent' div
    for tag in body_content.find_all(['p', 'h2'], recursive=True):
        if tag.name == 'h2':
            break
        if not_in_table(tag):
            summary_text += tag.text
    
    if summary_text:
        print("Scraped Summary:")
        print(summary_text)
    else:
        print("No summary text found.")
```
